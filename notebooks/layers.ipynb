```python
from matilda.layers import HarmonicTransform
from matilda.transform import show_filter_bank

import numpy as np
import tensorflow as tf
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = data.astronaut()
```

# Harmonic transformation layer

```python
input = tf.keras.layers.Input(shape=image.shape)
layer = HarmonicTransform('dct', n=3)
output = layer(input)
model = tf.keras.models.Model(inputs=input, outputs=output)
```

```python
show_filter_bank(layer.filter_bank.numpy())
```

```python
result = model(image[np.newaxis, ...])
for i in range(27):
    plt.imshow(result[:, :, :, i].numpy().squeeze())
    plt.show()
```
